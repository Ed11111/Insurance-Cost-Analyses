```python
import csv
with open("insurance.csv") as ins_csv:    
    patients = list(csv.DictReader(ins_csv))
    
    # The i is made the key so that we don't overwrite patients with the same ages. 
    # I.e. if we made the age the key, people of the same age will overwrite each other in the dictionary.
    ages_n_costs = {i:{"Age":int(patients[i]["age"]),"Charges":float(patients[i]["charges"]),"Smoker": patients[i]["smoker"]} 
                        for i in range(len(patients))
                       }
    
    # This sets up the boundries we want to aggregate the data into.
    age_boundries = [18,20,25,30,35,40,45,50,55,60,65,70,75,80] 
    age_boundries_index = len(age_boundries) - 1   
    
    # This sets up the the dictionaries we use to keep track of the totals of the charges and the counts of the patients. 
    ages_n_charges_total = {}
    ages_n_charges_smokers_total = {}
    
    patient_count = {}
    patient_count_smokers = {}
    
    for patients in ages_n_costs.values():
        
        #This for statement will go through each of the age boundries set by the age_boundry above. 
        for n in range(0,age_boundries_index):
            
            # This variable defines how the boundry is written in the key.
            boundry = f"{age_boundries[n]} <= age < {age_boundries[n+1]}"
            
            # This if statement will choose non smokers within an age range set by the age_boundry above.
            if patients["Smoker"] == "no" and patients["Age"] >= age_boundries[n] and patients["Age"] < age_boundries[n+1]:
                
                    #The dictionaries are written and appended the same way we did in the Hurricane analyses project.
                    if boundry not in ages_n_charges_total and boundry not in patient_count:
                        ages_n_charges_total[boundry] = float(patients["Charges"])
                        patient_count[boundry] = 1
                    else:
                        ages_n_charges_total[boundry] += float(patients["Charges"])
                        patient_count[boundry] += 1
                        
            # This if statement will choose smokers within an age range set by the age_boundry above.
            elif patients["Smoker"] == "yes" and patients["Age"] >= age_boundries[n] and patients["Age"] < age_boundries[n+1]:
                    if boundry not in ages_n_charges_smokers_total and boundry not in patient_count_smokers:
                        ages_n_charges_smokers_total[boundry] = float(patients["Charges"])
                        patient_count_smokers[boundry] = 1
                    else:
                        ages_n_charges_smokers_total[boundry] += float(patients["Charges"])
                        patient_count_smokers[boundry] += 1
    
    # These are dictionaries of the calculated averages and their sorted versions.
    ages_n_charges_average = {boundry:round(ages_n_charges_total[boundry]/patient_count[boundry],2) for boundry in ages_n_charges_total}
    sorted_ages_n_charges_average = dict(sorted(ages_n_charges_average.items())) 
    ages_n_charges_average_smokers = {boundry:round(ages_n_charges_smokers_total[boundry]/patient_count_smokers[boundry],2) for boundry in ages_n_charges_smokers_total}
    sorted_ages_n_charges_average_smokers = dict(sorted(ages_n_charges_average_smokers.items()))
    
    print(f"Non-smoker averages by age:\n{sorted_ages_n_charges_average}")
    print(f"Non-smoker patient count by age:\n{dict(sorted(patient_count.items()))}")
    print(f"Smoker averages by age: \n{sorted_ages_n_charges_average_smokers}")
    print(f"Smoker patient count by age:\n{dict(sorted(patient_count_smokers.items()))}")
    
    # Rinse and repeat for the other data categories.
   
```

```python
import csv
with open("insurance.csv",) as ins_csv:    
    patients = list(csv.DictReader(ins_csv))      
    region_n_costs = {i:{"Region":str(patients[i]["region"]),"Charges":float(patients[i]["charges"]),"Smoker": patients[i]["smoker"]}
                     for i in range(len(patients))
                     }                   
    
    patient_count = {}
    patient_count_smokers = {}
    regions = {patient["Region"] for patient in region_n_costs.values()}
    print(regions)
    regions_total_costs = {}
    regions_total_costs_smokers = {}
    for patient in region_n_costs.values():
        for region in regions:
            if patient["Smoker"] == "no" and patient["Region"] == region:
                if region not in regions_total_costs and region not in patient_count:
                      regions_total_costs[region] = float(patient["Charges"])
                      patient_count[region] = 1
                else:
                    regions_total_costs[region] += float(patient["Charges"])
                    patient_count[region] += 1
            elif patient["Smoker"] == "yes" and patient["Region"] == region:
                if region not in regions_total_costs_smokers and region not in patient_count_smokers:
                    regions_total_costs_smokers[region] = float(patient["Charges"])
                    patient_count_smokers[region] = 1
                else:
                    regions_total_costs_smokers[region] += float(patient["Charges"])
                    patient_count_smokers[region] += 1
    regions_average_costs = {region:round(regions_total_costs[region]/patient_count[region],2) for region in regions_total_costs}
    sorted_regions_average_costs = dict(sorted(regions_average_costs.items()))
    regions_average_costs_smokers = {region:round(regions_total_costs_smokers[region]/patient_count_smokers[region],2) for region in regions_total_costs_smokers}
    sorted_regions_average_costs_smoker = dict(sorted(regions_average_costs_smokers.items()))
    print(f"Non-smoker averages by region:\n{sorted_regions_average_costs}")
    print(f"Non-smoker patient count by region:\n{dict(sorted(patient_count.items()))}")
    print(f"Smoker averages by region:\n{sorted_regions_average_costs_smoker}")
    print(f"Smoker patient count by region:\n{dict(sorted(patient_count_smokers.items()))}")
```

```python
import csv
with open("insurance.csv",) as ins_csv:    
    patients = list(csv.DictReader(ins_csv))         
    children_n_costs = {i:{"Children":int(patients[i]["children"]),"Charges":float(patients[i]["charges"]),"Smoker": patients[i]["smoker"]}
                    for i in range(len(patients))
                   }

    max_children = 0
    for patient in children_n_costs.values(): 
        if patient["Children"] > max_children: 
            max_children = patient["Children"] 
    children_boundries = list(range(0,max_children+1))
    print(f"Children numbers: {children_boundries}")
    children_boundries_index = len(children_boundries)   

    children_n_charges_total = {}
    children_n_charges_smokers_total = {}

    patient_count = {}
    patient_count_smokers = {}

    for patients in children_n_costs.values():       
        for n in range(0,children_boundries_index):           
            boundry = f"{children_boundries[n]}"
            if patients["Smoker"] == "no" and patients["Children"] == children_boundries[n]:
                    if boundry not in children_n_charges_total and boundry not in patient_count:
                        children_n_charges_total[boundry] = float(patients["Charges"])
                        patient_count[boundry] = 1
                    else:
                        children_n_charges_total[boundry] += float(patients["Charges"])
                        patient_count[boundry] += 1
            elif patients["Smoker"] == "yes" and patients["Children"] == children_boundries[n]:
                    if boundry not in children_n_charges_smokers_total and boundry not in patient_count_smokers:
                        children_n_charges_smokers_total[boundry] = float(patients["Charges"])
                        patient_count_smokers[boundry] = 1
                    else:
                        children_n_charges_smokers_total[boundry] += float(patients["Charges"])
                        patient_count_smokers[boundry] += 1
 
    children_n_charges_average = {boundry:round(children_n_charges_total[boundry]/patient_count[boundry],2) for boundry in children_n_charges_total}
    sorted_children_n_charges_average = dict(sorted(children_n_charges_average.items()))
    children_n_charges_average_smokers = {boundry:round(children_n_charges_smokers_total[boundry]/patient_count_smokers[boundry],2) for boundry in children_n_charges_smokers_total}
    sorted_children_n_charges_average_smokers = dict(sorted(children_n_charges_average_smokers.items()))
    print(f"Non-smoker charges averages by children: \n{sorted_children_n_charges_average}")
    print(f"Non-smoker patient count by number of children:\n{dict(sorted(patient_count.items()))}")
    print(f"Smoker charges averages by children: \n{sorted_children_n_charges_average_smokers}")
    print(f"Smoker patient count by number of children:\n{dict(sorted(patient_count_smokers.items()))}")
```
